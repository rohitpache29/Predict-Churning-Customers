```python
cat1 = df.select_dtypes(exclude = ['int64','float64']).columns
int1 = df.select_dtypes(include = ['int64']).columns
float1 = df.select_dtypes(include = ['Float64']).columns


df = pd.get_dummies(df, columns = cat1)


from sklearn.preprocessing import MinMaxScaler
for col in df[float1]:
    df[col] = MinMaxScaler().fit_transform(df[[col]])

for col in df[int1]:
    df[col] = MinMaxScaler().fit_transform(df[[col]])

print('New Number of Features: %d'%(df.shape[1]))
```

```python
corr1 = df.corr()
corr1.sort_values(["Attrition_Flag"], ascending = False, inplace = True)
correlations2 = corr1.Attrition_Flag

a = correlations2[correlations2 > 0.1]
b = correlations2[correlations2 < -0.1]

top_corr_features = a.append(b)

top_corr_features
print(top_corr_features)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
#preparing data for training
x=df[['Contacts_Count_12_mon','Months_Inactive_12_mon','Total_Amt_Chng_Q4_Q1','Total_Relationship_Count','Total_Trans_Amt','Avg_Utilization_Ratio','Total_Revolving_Bal','Total_Ct_Chng_Q4_Q1','Total_Trans_Ct']]
y=df['Attrition_Flag']
```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=RandomForestClassifier(n_estimators=100)
model.fit(x_train,y_train)
```

```python
model_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,model_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model2=LogisticRegression().fit(x_train,y_train)
```

```python
y_test_pred=model2.predict(x_test)
```

```python
print(classification_report(y_test,y_test_pred))
```

Random forest classifier model is better than logistic regression model because of better accuracy
